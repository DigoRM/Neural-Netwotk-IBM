## Import Necessary Libraries

```python
import pandas as pd
import numpy as np

import keras
from keras.models import Sequential
from keras.layers import Dense
```

## Import Data

```python
df = pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv")
df.head()
```

## Data info

```python
df.isnull().sum()
```

## Split Predictors and Target values

```python
concrete_data_columns = df.columns

X = df.drop('Strength', 1).values # all columns except Strength
y = df['Strength'] # Strength column
```

## Normalize Data (new!!)

```python
X_norm = (X - X.mean()) / X.std()
```

## Split data into Train and Test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size = 0.3, random_state=6)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Build and Train Neural Network

```python
n_cols = X.shape[1]
n_cols
```

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    # model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# build the model
model = regression_model()

# train
np.random.seed(45)
model.fit(X_train, y_train, epochs=50, verbose=1)
```

## Predict and Evaluate

```python
predictions = model.predict(X_test)
print(predictions.shape)
print(y_test.shape)
```

```python
predictions = predictions[0::]
y_test = np.asarray(y_test)
y_test
```

```python
predictions
```

```python
print("Mean absolute error: %.2f" % np.mean(np.absolute(predictions - y_test)))
print("Residual sum of squares (MSE): %.2f" % np.mean((predictions - y_test) ** 2))
from sklearn.metrics import r2_score
print("R2-score: %.2f" % r2_score(y_test, predictions) )
```
